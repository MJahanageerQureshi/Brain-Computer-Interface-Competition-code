```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt

data = sio.loadmat('bci_comp.mat')

x = data['x']
y = data['y']

x_scale = preprocessing.scale(x)
plt.plot(x_scale)
plt.show()
```

```python
from scipy.signal import medfilt
x_med = medfilt(x_scale)

plt.plot(x_scale)
plt.show()
plt.plot(x_med)
plt.show()
```

```python
right = []
left = []
for i in range(np.shape(x_scale)[0]):
    if y[i] == 1:
        right.append(x_scale[i])
    else:
        left.append(x_scale[i])

plt.plot(right)
plt.show()
plt.plot(left)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as qda
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as lda
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

x_scale = x_med

knn = KNeighborsClassifier()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
results = cross_val_score(knn, x_scale, y, cv=kfold)
print('knn')
print(results)
print(results.mean())

lda = lda()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
results = cross_val_score(lda, x_scale, y, cv=kfold)
print('lda')
print(results)
print(results.mean())

qda = qda()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
results = cross_val_score(qda, x_scale, y, cv=kfold)
print('qda')
print(results)
print(results.mean())

gb = GaussianNB()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
results = cross_val_score(gb, x_scale, y, cv=kfold)
print('gb')
print(results)
print(results.mean())

svm = SVC()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
results = cross_val_score(svm, x_scale, y, cv=kfold)
print('svm')
print(results)
print(results.mean())

dt = DecisionTreeClassifier()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
results = cross_val_score(dt, x_scale, y, cv=kfold)
print('dt')
print(results)
print(results.mean())

rf = RandomForestClassifier()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
results = cross_val_score(rf, x_scale, y, cv=kfold)
print('rf')
print(results)
print(results.mean())

adb = AdaBoostClassifier()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
results = cross_val_score(adb, x_scale, y, cv=kfold)
print('adb')
print(results)
print(results.mean())
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers.normalization import BatchNormalization

def feature_extractor():
    model = Sequential()
    model.add(Dense(100, input_dim=24, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(50, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(10, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(.3))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model = feature_extractor()
X_train, X_test, Y_train, Y_test = train_test_split(x_scale, y, random_state = 42, test_size = .1)
model.fit(X_train, Y_train, epochs=200, batch_size=5, validation_split = .3, verbose = 1)
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
model.summary()
```

```python
from keras.models import Model

layer_name = 'batch_normalization_49'
automatic_feature_extraction = Model(inputs=model.input,outputs=model.get_layer(layer_name).output)

automatic_feature_extraction.summary()
```

```python
x_features = automatic_feature_extraction.predict(x_med)
print(np.shape(x_features))

from sklearn import preprocessing
x_scale_features = preprocessing.scale(x_features)

import matplotlib.pyplot as plt
plt.plot(x_features)
plt.show()
plt.plot(x_scale_features)
plt.show()
```
